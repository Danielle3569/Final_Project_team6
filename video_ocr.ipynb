```python
import cv2
import pytesseract
import os

# Tesseract 경로 설정 (Windows 예시)
pytesseract.pytesseract.tesseract_cmd = r'C:\Users\USER\AppData\Local\Programs\Tesseract-OCR\tesseract.exe'

def preprocess_image(image):
    # 그레이스케일로 변환
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # 대비를 높이기 위해 이진화
    _, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
    # 노이즈 제거
    processed_image = cv2.medianBlur(binary, 3)
    return processed_image

def extract_text_from_video(video_path, frame_interval=30):
    # 동영상 파일 열기
    cap = cv2.VideoCapture(video_path)
    
    # 동영상이 성공적으로 열렸는지 확인
    if not cap.isOpened():
        print("Error: Could not open video.")
        return
    
    frame_count = 0
    extracted_text = ""
    
    while True:
        # 동영상에서 프레임 읽기
        ret, frame = cap.read()
        
        # 프레임을 성공적으로 읽지 못하면 루프 종료
        if not ret:
            break
        
        # 'frame_interval' 프레임마다 처리
        if frame_count % frame_interval == 0:
            # 이미지 전처리
            processed_frame = preprocess_image(frame)
            
            # 임시 파일 경로 설정
            temp_dir = os.path.join(os.getcwd(), 'temp')
            if not os.path.exists(temp_dir):
                os.makedirs(temp_dir)
            temp_file = os.path.join(temp_dir, 'temp_image.png')
            cv2.imwrite(temp_file, processed_frame)
            
            # pytesseract를 사용하여 프레임에서 텍스트 추출
            text = pytesseract.image_to_string(temp_file, config='--psm 6')
            extracted_text += text + "\n"
        
        frame_count += 1
    
    # 동영상 캡처 객체 해제
    cap.release()
    
    return extracted_text

# 사용 예시
video_path = "videoplayback (1).mp4"  # 다운로드한 동영상 파일 경로
text = extract_text_from_video(video_path)
print(text)
```
